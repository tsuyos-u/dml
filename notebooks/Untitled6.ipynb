```python
import numpy as np
import matplotlib.pyplot as plt
import mat

# 設定
np.random.seed(42)
n_points = 10  # 各個体の点数
n_individuals = 3  # 個体数

# 色の設定
colors = ['red', 'blue', 'green']
labels = ['個体A', '個体B', '個体C']

# 図の作成
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# === 1. 固定効果（切片とXが相関あり） ===
x1 = np.random.uniform(1, 10, n_points)
x2 = np.random.uniform(3, 12, n_points)  
x3 = np.random.uniform(5, 14, n_points)

# 切片をXの平均と相関させる
intercept1 = np.mean(x1) * 0.5 + 2  # X平均が小さい→切片小さい
intercept2 = np.mean(x2) * 0.5 + 2  # X平均が中程度→切片中程度
intercept3 = np.mean(x3) * 0.5 + 2  # X平均が大きい→切片大きい

y1 = intercept1 + 1.2 * x1 + np.random.normal(0, 1, n_points)
y2 = intercept2 + 1.2 * x2 + np.random.normal(0, 1, n_points)
y3 = intercept3 + 1.2 * x3 + np.random.normal(0, 1, n_points)

# プロット
ax1.scatter(x1, y1, color=colors[0], label=labels[0], s=60)
ax1.scatter(x2, y2, color=colors[1], label=labels[1], s=60)
ax1.scatter(x3, y3, color=colors[2], label=labels[2], s=60)

ax1.set_title('固定効果\n（切片とXが相関あり）', fontsize=14)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.legend()
ax1.grid(True, alpha=0.3)

# === 2. ランダム効果（切片とXが相関なし） ===
x1_r = np.random.uniform(1, 10, n_points)
x2_r = np.random.uniform(1, 10, n_points)  # 同じ範囲
x3_r = np.random.uniform(1, 10, n_points)  # 同じ範囲

# 切片をランダムに設定（Xと無関係）
intercept1_r = 3
intercept2_r = 7  
intercept3_r = 5

y1_r = intercept1_r + 1.2 * x1_r + np.random.normal(0, 1, n_points)
y2_r = intercept2_r + 1.2 * x2_r + np.random.normal(0, 1, n_points)
y3_r = intercept3_r + 1.2 * x3_r + np.random.normal(0, 1, n_points)

# プロット
ax2.scatter(x1_r, y1_r, color=colors[0], label=labels[0], s=60)
ax2.scatter(x2_r, y2_r, color=colors[1], label=labels[1], s=60)
ax2.scatter(x3_r, y3_r, color=colors[2], label=labels[2], s=60)

ax2.set_title('ランダム効果\n（切片とXが相関なし）', fontsize=14)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("固定効果：Xが大きい個体ほど切片も大きい")
print("ランダム効果：Xと切片は無関係")
```
